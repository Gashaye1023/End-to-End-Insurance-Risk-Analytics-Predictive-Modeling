```python
from scipy.stats import chi2_contingency, ttest_ind, f_oneway
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
from scipy import stats
```

```python

# Load and prepare data
try:
    df = pd.read_csv('MachineLearningRating_v3_processed.csv')
    
    # Check for required columns
    required_columns = ['TotalPremium', 'TotalClaims', 'Province', 'PostalCode', 'Gender']
    missing_cols = [col for col in required_columns if col not in df.columns]
    
    if missing_cols:
        raise KeyError(f"Missing required columns: {missing_cols}")
        
    # Create necessary metrics
    df['Claimed'] = (df['TotalClaims'] > 0).astype(int)
    df['Margin'] = df['TotalPremium'] - df['TotalClaims']
    df_claims = df[df['Claimed'] == 1].copy()
    
except Exception as e:
    print(f"Error loading or preparing data: {e}")
    # Create sample data if real data fails
    print("Creating sample data for demonstration...")
    np.random.seed(42)
    data = {
        'Province': np.random.choice(['Gauteng', 'Western Cape', 'KZN'], size=1000),
        'PostalCode': np.random.randint(1000, 9999, size=1000),
        'Gender': np.random.choice(['Male', 'Female'], size=1000),
        'TotalPremium': np.random.uniform(100, 5000, size=1000),
        'TotalClaims': np.random.uniform(0, 3000, size=1000)
    }
    df = pd.DataFrame(data)
    df['Claimed'] = (df['TotalClaims'] > 0).astype(int)
    df['Margin'] = df['TotalPremium'] - df['TotalClaims']
    df_claims = df[df['Claimed'] == 1].copy()

def hypothesis_test(df, group_var, metric, test_type='t-test', top_n=5):
    print(f"\n=== Testing {metric} by {group_var} ===")
    
    # Check if columns exist
    if group_var not in df.columns:
        print(f"Error: Grouping variable '{group_var}' not found in DataFrame")
        return
    if metric not in df.columns and metric != 'Claimed':
        print(f"Error: Metric '{metric}' not found in DataFrame")
        return
    
    # Get top groups
    top_groups = df[group_var].value_counts().nlargest(top_n).index
    df_filtered = df[df[group_var].isin(top_groups)]
    
    if test_type == 'chi2' and metric == 'Claimed':
        # Claim frequency test
        try:
            contingency = pd.crosstab(df_filtered[group_var], df_filtered['Claimed'])
            if contingency.shape[0] > 1 and contingency.shape[1] > 1:
                chi2, p_val, _, _ = chi2_contingency(contingency)
                print(f"Chi-square p-value: {p_val:.4f}")
                if p_val < 0.05:
                    print("Significant differences found:")
                    print(df_filtered.groupby(group_var)['Claimed'].mean().sort_values(ascending=False))
                else:
                    print("No significant differences in claim frequency")
            else:
                print("Not enough data for chi-square test")
        except Exception as e:
            print(f"Error performing chi-square test: {e}")
    else:
        # For continuous metrics (severity or margin)
        try:
            groups = df_filtered.groupby(group_var)[metric]
            if len(groups) > 1:
                group_data = [group[1][metric].values for group in groups]
                
                if len(groups) == 2:
                    # T-test for 2 groups
                    t_stat, p_val = ttest_ind(*group_data, equal_var=False)
                    print(f"T-test p-value: {p_val:.4f}")
                else:
                    # ANOVA for >2 groups
                    f_stat, p_val = f_oneway(*group_data)
                    print(f"ANOVA p-value: {p_val:.4f}")
                
                if p_val < 0.05:
                    print("Significant differences found:")
                    print(groups.mean().sort_values(ascending=False))
                else:
                    print("No significant differences")
            else:
                print("Not enough groups for comparison")
        except Exception as e:
            print(f"Error performing {test_type}: {e}")
    
    # Visualization
    try:
        plt.figure(figsize=(10, 5))
        if metric == 'Claimed':
            sns.barplot(x=group_var, y=metric, data=df_filtered, estimator=np.mean)
            plt.title(f'Claim Frequency by {group_var}')
        else:
            sns.boxplot(x=group_var, y=metric, data=df_filtered)
            plt.title(f'{metric} by {group_var}')
        plt.xticks(rotation=45)
        plt.tight_layout()
        plt.show()
    except Exception as e:
        print(f"Error creating visualization: {e}")

# Test all hypotheses
print("=== Hypothesis Testing Results ===")
hypothesis_test(df, 'Province', 'Claimed', 'chi2')  # H1: Risk differences across provinces (frequency)
hypothesis_test(df_claims, 'Province', 'TotalClaims', 'anova')  # H1: Risk differences (severity)
hypothesis_test(df, 'Province', 'Margin', 'anova')  # H1: Margin differences

hypothesis_test(df, 'PostalCode', 'Claimed', 'chi2')  # H2: Risk differences by zip (frequency)
hypothesis_test(df_claims, 'PostalCode', 'TotalClaims', 'anova')  # H2: Risk differences (severity)
hypothesis_test(df, 'PostalCode', 'Margin', 'anova')  # H3: Margin differences by zip

hypothesis_test(df, 'Gender', 'Claimed', 'chi2')  # H4: Risk differences by gender (frequency)
hypothesis_test(df_claims, 'Gender', 'TotalClaims', 't-test')  # H4: Risk differences (severity)
hypothesis_test(df, 'Gender', 'Margin', 't-test')  # H4: Margin differences by gender
```

```python
from scipy.stats import chi2_contingency, ttest_ind, f_oneway

# Ensure 'Claimed' column is available
if 'Claimed' not in df.columns:
    df['Claimed'] = (df['TotalClaims'] > 0).astype(int)

# Filter for claims > 0 for Claim Severity calculation
df_claims_only = df[df['Claimed'] == 1].copy()
province_claim_freq = pd.crosstab(df['Province'], df['Claimed'])

# Initialize p_val
p_val = None

if not province_claim_freq.empty and province_claim_freq.shape[0] > 1 and province_claim_freq.shape[1] > 1:
    chi2, p_val, _, _ = chi2_contingency(province_claim_freq)
    print(f"Claim Frequency (Province) - Chi-squared test p-value: {p_val:.4f}")

    if p_val < 0.05:
        print(df.groupby('Province')['Claimed'].mean().sort_values(ascending=False).head())
else:
    print("Not enough data to perform Chi-squared test for Province vs Claimed.")

# Claim Severity by Province (ANOVA test)
provinces = df_claims_only['Province'].unique()
if len(provinces) > 1: 
    province_claim_amounts = [df_claims_only[df_claims_only['Province'] == p]['TotalClaims'] for p in provinces]

    # Filter out empty arrays for ANOVA
    province_claim_amounts = [arr for arr in province_claim_amounts if len(arr) > 0]

    if len(province_claim_amounts) > 1:
        f_stat, p_val = f_oneway(*province_claim_amounts)
        print(f"Claim Severity (Province) - ANOVA test p-value: {p_val:.4f}")

        if p_val < 0.05:
            print(df_claims_only.groupby('Province')['TotalClaims'].mean().sort_values(ascending=False).head())
        else:
            print("Conclusion: Fail to Reject the Null Hypothesis. No significant risk (claim severity) differences across provinces.")
    else:
        print("Not enough provinces with claims to perform ANOVA test.")
```

```python
# Train-Test Split for Claim Severity Prediction
categorical_features = df_claims.select_dtypes(include=['object']).columns.tolist()
numerical_features = df_claims.select_dtypes(include=['number']).columns.tolist()
numerical_features = [col for col in numerical_features if col != 'TotalClaims']
X = df_claims[categorical_features + numerical_features]
y = df_claims['TotalClaims']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
# Step 4: Define Preprocessing and Modeling Pipelines
# Preprocessing pipeline
preprocessor = ColumnTransformer(
    transformers=[
        ('num', 'passthrough', numerical_features),
        ('cat', OneHotEncoder(), categorical_features)
    ]
)

# Define regression models
regression_models = {
    'Linear Regression': Pipeline(steps=[('preprocessor', preprocessor), ('model', LinearRegression())]),
    'Random Forest': Pipeline(steps=[('preprocessor', preprocessor), ('model', RandomForestRegressor())]),
    'XGBoost': Pipeline(steps=[('preprocessor', preprocessor), ('model', xgb.XGBRegressor())])
}
```

```python
# Fit and evaluate regression models
regression_results = {}
for name, model in regression_models.items():
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    rmse = np.sqrt(mean_squared_error(y_test, y_pred))
    r2 = r2_score(y_test, y_pred)
    regression_results[name] = {'RMSE': rmse, 'R2': r2}

# Report regression results
for model_name, metrics in regression_results.items():
    print(f"{model_name}: RMSE = {metrics['RMSE']:.2f}, R² = {metrics['R2']:.2f}")
```

```python
# Step 5: SHAP analysis for the best regression model
best_regression_model = regression_models['XGBoost']
explainer = shap.Explainer(best_regression_model.named_steps['model'])
shap_values = explainer(best_regression_model.named_steps['preprocessor'].transform(X_train))
shap.summary_plot(shap_values, features=X_train)
```
