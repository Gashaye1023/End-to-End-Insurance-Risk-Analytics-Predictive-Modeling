```python
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings("ignore")
```

```python
file_path = 'MachineLearningRating_v3.txt'

# Define the column names
columns = [
    'UnderwrittenCoverID', 'PolicyID', 'TransactionMonth',
    'IsVATRegistered', 'Citizenship', 'LegalType', 'Title',
    'Language', 'Bank', 'AccountType', 'MaritalStatus', 'Gender',
    'Country', 'Province', 'PostalCode', 'MainCrestaZone',
    'SubCrestaZone', 'ItemType', 'Mmcode', 'VehicleType',
    'RegistrationYear', 'Make', 'Model', 'Cylinders',
    'Cubiccapacity', 'Kilowatts', 'Bodytype', 'NumberOfDoors',
    'VehicleIntroDate', 'CustomValueEstimate', 'AlarmImmobiliser',
    'TrackingDevice', 'CapitalOutstanding', 'NewVehicle',
    'WrittenOff', 'Rebuilt', 'Converted', 'CrossBorder',
    'NumberOfVehiclesInFleet', 'SumInsured', 'TermFrequency',
    'CalculatedPremiumPerTerm', 'ExcessSelected', 'CoverCategory',
    'CoverType', 'CoverGroup', 'Section', 'Product',
    'StatutoryClass', 'StatutoryRiskType', 'TotalPremium',
    'TotalClaims'
]

# Read the data from the text file
df = pd.read_csv(file_path, sep='|', header=None, names=columns, na_values=[' '])

# Save the DataFrame to a CSV file
output_csv_path = 'MachineLearningRating_v3.csv'
df.to_csv(output_csv_path, index=False)
```

```python
    # 1.4 Bivariate or Multivariate Analysis
print("\n Bivariate/Multivariate Analysis:")

    # Loss Ratio by Province, VehicleType, and Gender
print("\nLoss Ratio by Province:")
province_loss_ratio = df.groupby('Province').apply(lambda x: x['TotalClaims'].sum() / x['TotalPremium'].sum() * 100 if x['TotalPremium'].sum() > 0 else 0).sort_values(ascending=False)
print(province_loss_ratio)
plt.figure(figsize=(10, 6))
sns.barplot(x=province_loss_ratio.index, y=province_loss_ratio.values, palette='viridis')
plt.title('Loss Ratio by Province')
plt.xlabel('Province')
plt.ylabel('Loss Ratio (%)')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()

print("\nLoss Ratio by VehicleType:")
vehicletype_loss_ratio = df.groupby('VehicleType').apply(lambda x: x['TotalClaims'].sum() / x['TotalPremium'].sum() * 100 if x['TotalPremium'].sum() > 0 else 0).sort_values(ascending=False)
print(vehicletype_loss_ratio)
plt.figure(figsize=(10, 6))
sns.barplot(x=vehicletype_loss_ratio.index, y=vehicletype_loss_ratio.values, palette='magma')
plt.title('Loss Ratio by Vehicle Type')
plt.xlabel('Vehicle Type')
plt.ylabel('Loss Ratio (%)')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()

print("\nLoss Ratio by Gender:")
gender_loss_ratio = df.groupby('Gender').apply(lambda x: x['TotalClaims'].sum() / x['TotalPremium'].sum() * 100 if x['TotalPremium'].sum() > 0 else 0).sort_values(ascending=False)
print(gender_loss_ratio)
plt.figure(figsize=(6, 4))
sns.barplot(x=gender_loss_ratio.index, y=gender_loss_ratio.values, palette='plasma')
plt.title('Loss Ratio by Gender')
plt.xlabel('Gender')
plt.ylabel('Loss Ratio (%)')
plt.tight_layout()
plt.show()
```

```python
    # Temporal Trends: Claim Frequency and Severity over time
    print("\nTemporal Trends: Claim Frequency and Severity over 18-month period:")
    monthly_data = df.groupby('TransactionMonth').agg(
        TotalPolicies=('PolicyID', 'count'),
        TotalClaimsSum=('TotalClaims', 'sum'),
        PoliciesWithClaims=('Claimed', 'sum')
    ).reset_index()

    monthly_data['ClaimFrequency'] = (monthly_data['PoliciesWithClaims'] / monthly_data['TotalPolicies']) * 100
    monthly_data['ClaimSeverity'] = (monthly_data['TotalClaimsSum'] / monthly_data['PoliciesWithClaims']).fillna(0) # Severity only for policies with claims

    plt.figure(figsize=(14, 6))
    plt.subplot(1, 2, 1)
    sns.lineplot(x='TransactionMonth', y='ClaimFrequency', data=monthly_data, marker='o')
    plt.title('Monthly Claim Frequency')
    plt.xlabel('Month')
    plt.ylabel('Claim Frequency (%)')
    plt.xticks(rotation=45)
    plt.grid(True, linestyle='--', alpha=0.7)

    plt.subplot(1, 2, 2)
    sns.lineplot(x='TransactionMonth', y='ClaimSeverity', data=monthly_data, marker='o')
    plt.title('Monthly Claim Severity (Avg. Claim Amount)')
    plt.xlabel('Month')
    plt.ylabel('Claim Severity (Rand)')
    plt.xticks(rotation=45)
    plt.grid(True, linestyle='--', alpha=0.7)

    plt.tight_layout()
    plt.show()
```

```python
    # Correlations between TotalPremium, TotalClaims, and PostalCode
    # Due to large number of PostalCodes, aggregate by PostalCode
    postal_code_agg = df.groupby('PostalCode').agg(
        AvgTotalPremium=('TotalPremium', 'mean'),
        AvgTotalClaims=('TotalClaims', 'mean'),
        NumPolicies=('PolicyID', 'count')
    ).reset_index()

    # Filter for postal codes with a reasonable number of policies for meaningful correlation
    postal_code_agg = postal_code_agg[postal_code_agg['NumPolicies'] >= 50] # Threshold for meaningful analysis

    if not postal_code_agg.empty:
        plt.figure(figsize=(10, 7))
        sns.scatterplot(x='AvgTotalPremium', y='AvgTotalClaims', data=postal_code_agg, hue='NumPolicies', size='NumPolicies', sizes=(20, 400), alpha=0.6)
        plt.title('Avg. Total Claims vs Avg. Total Premium by Postal Code')
        plt.xlabel('Average Total Premium')
        plt.ylabel('Average Total Claims')
        plt.xscale('log')
        plt.yscale('log')
        plt.grid(True, linestyle='--', alpha=0.5)
        plt.tight_layout()
        plt.show()
        print("\nCorrelation between Average Total Premium and Average Total Claims by Postal Code:")
        print(postal_code_agg[['AvgTotalPremium', 'AvgTotalClaims']].corr())
    else:
        print("\nNot enough unique Postal Codes with sufficient data to plot meaningful correlations.")


    # 1.5 Outlier Detection: Box plots for numerical data
    print("\n1.5 Outlier Detection: Box Plots:")
    plt.figure(figsize=(15, 8))
    plt.suptitle('Outlier Detection for Key Financial Variables', fontsize=16)

    plt.subplot(1, 3, 1)
    sns.boxplot(y=df['TotalPremium'])
    plt.title('Box Plot of TotalPremium')
    plt.ylabel('TotalPremium')

    plt.subplot(1, 3, 2)
    sns.boxplot(y=df['TotalClaims'])
    plt.title('Box Plot of TotalClaims')
    plt.ylabel('TotalClaims')

    plt.subplot(1, 3, 3)
    sns.boxplot(y=df['CustomValueEstimate'])
    plt.title('Box Plot of CustomValueEstimate')
    plt.ylabel('CustomValueEstimate')

    plt.tight_layout(rect=[0, 0.03, 1, 0.95])
    plt.show()

    # 1.6 Visualization: 3 Creative and Beautiful Plots
    print("\n1.6 Creative Visualizations:")

    # Plot 1: Premium vs. Claims by Gender and CoverCategory
    plt.figure(figsize=(12, 7))
    sns.scatterplot(x='TotalPremium', y='TotalClaims', hue='Gender', style='CoverCategory',
                    data=df.sample(n=min(5000, len(df)), random_state=42), # Sample for better visualization of large datasets
                    alpha=0.6, s=100)
    plt.title('Total Premium vs. Total Claims by Gender and Cover Category')
    plt.xlabel('Total Premium')
    plt.ylabel('Total Claims')
    plt.xscale('log') # Use log scale for better distribution visibility
    plt.yscale('log')
    plt.grid(True, which="both", ls="--", c='0.7')
    plt.legend(title='Gender & Cover Category', bbox_to_anchor=(1.05, 1), loc='upper left')
    plt.tight_layout()
    plt.show()

    # Plot 2: Claim Frequency by Vehicle Type and Province
    claim_frequency_pivot = df.groupby(['VehicleType', 'Province'])['Claimed'].mean().unstack()
    plt.figure(figsize=(14, 8))
    sns.heatmap(claim_frequency_pivot, annot=True, fmt=".2%", cmap="YlGnBu", linewidths=.5, cbar_kws={'label': 'Claim Frequency'})
    plt.title('Claim Frequency (%) Heatmap by Vehicle Type and Province')
    plt.xlabel('Province')
    plt.ylabel('Vehicle Type')
    plt.tight_layout()
    plt.show()

```
